# 공공 인공지능 Open API 활용

### 객체 검출

```python
import urllib3, json, base64
openApiURL = "http://aiopen.etri.re.kr:8000/ObjectDetect"
```

- 여러가지 객체 검출

```python
with open('etriaikey.txt') as f:
    ai_key = f.read()
```

```python
img_file = '사진/공원.jpg'
img_type = img_file.split('.')[-1]
img_type
```

```python
import matplotlib as mpl
mpl.rcParams['axes.unicode_minus'] = False
plt.rcParams['font.family'] = 'Malgun Gothic'
```

```python
# JFIF 타입은 JPG로 변경
img_type = 'jpg' if img_type == 'jfif' else img_type
```

```python
with open(img_file, "rb") as file:
    img_contents = base64.b64encode(file.read()).decode("utf8")
```

```python
request_json = {
    "access_key": ai_key,
    "argument": {
        "type": img_type,
        "file": img_contents
    }
}
```

```python
# 내가 보낼 데이터는 제이슨 타입, 스트링으로 포장해서 
http = urllib3.PoolManager()
response = http.request(
    "POST",
    openApiURL,
    headers={"Content-Type": "application/json; charset=UTF-8"},
    body=json.dumps(request_json)
)
```

```python
response.status
```

```python
response.data
```

```python
result = json.loads(response.data)
obj_list = result['return_object']['data']
obj_list[0]
```

### - 시각화

```python
from PIL import Image, ImageDraw, ImageFont
import matplotlib.pyplot as plt
```

```python
img = Image.open(img_file)
plt.imshow(img);
```

```python
# pillow 에서 그림 그리기 # 사각형 그리고 글자 쓰겠다.
draw = ImageDraw.Draw(img)
for obj in obj_list:
    name = obj['class']

    x = int(obj['x'])   
    y = int(obj['y'])
    w = int(obj['width'])
    h = int(obj['height'])
    draw.rectangle(((x,y), (x+w, y+h)), outline=(255,0,0), width=2)
    draw.text((x+10, y-20), name, font=ImageFont.truetype('malgun.ttf', 10), fill=(0,0,255))
```

- 여러개의 그림에서 객체 검출하기

```python
# 글로브에 대해 알아야해 
from glob import glob
glob('사진/*.jpg')
```

```python
# http = urllib3.PoolManager() # 프로그램 안에서 한번만 실행하면 됨. 

for img_file in glob('사진/*jpg'):
    img_type = img_file.split('.')[-1]
    img_type = 'jpg' if img_type == 'jfif' else img_type
    with open(img_file, "rb") as file:
        img_contents = base64.b64encode(file.read()).decode("utf8")

    request_json = {
    "access_key": ai_key,
    "argument": {
        "type": img_type,
        "file": img_contents
        }
    }    
    response = http.request(
    "POST",
    openApiURL,
    headers={"Content-Type": "application/json; charset=UTF-8"},
    body=json.dumps(request_json)
    )
    if response.status !=200:
        continue
    
    result = json.loads(response.data)
    obj_list = result['return_object']['data']
    img = Image.open(img_file)
    draw = ImageDraw.Draw(img)
    for obj in obj_list:
        name = obj['class']
        x = int(obj['x'])   
        y = int(obj['y'])
        w = int(obj['width'])
        h = int(obj['height'])
        draw.rectangle(((x,y), (x+w, y+h)), outline=(255,0,0), width=2)
        draw.text((x+10, y-20), name, font=ImageFont.truetype('malgun.ttf', 20), fill=(255,0,0))

    plt.figure(figsize=(12,8))
    plt.imshow(img)
    plt.axis('off'), plt.title(img_file.split('\\')[-1])
    plt.show()
```
